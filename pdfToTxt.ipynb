```python
import textract
import glob
```

```python

for file in glob.glob('./**/*.pdf', recursive=True):
    write = False
    text = textract.process(file)
    with open('./txt/'+file.split('/')[-1].replace('pdf', 'txt'), 'a') as f:
        for page in text.decode().replace('\xa0', ' ').replace('\n\n', 'ʒ').replace('\n', ' ').replace('ʒ', '\n').split('\x0c')[1:]:
            if '1. Краткая информация' in page and 'Оглавление' not in page:
                write = True
            if 'Критерии оценки качества медицинской помощи' in page and 'Оглавление' not in page:
                write = False
                break
            if write:
                f.write(page)
    
```
